```python
import os
import shutil
import sys
import glob
from IPython.display import clear_output

#@markdown <font color=ORANGE>🔗 GitHub repository URL:</font>
GITHUB_URL = "https://github.com/Ali777666/VJ-Save-Restricted-Content-Colab"  #@param {type:"string"}

base_dir = './repo'

def clone_or_update_repo(repo_url, base_directory):
    repo_name = os.path.basename(repo_url).replace('.git', '')
    project_dir = os.path.join(base_directory, repo_name)
    if os.path.exists(project_dir):
        shutil.rmtree(project_dir)
    os.system(f'git clone {repo_url} {project_dir}')
    return project_dir

project_dir = clone_or_update_repo(GITHUB_URL, base_dir)
os.chdir(project_dir)

# Python versiya tekshiruvi (3.10-3.13)
version = sys.version_info
assert (3, 10) <= (version.major, version.minor) <= (3, 13), \
    f'Python 3.10-3.13 kerak, sizda: {version.major}.{version.minor}'
print(f'✅ Python {version.major}.{version.minor} — OK')

# Dependencies
os.system('pip install -r requirements.txt -q')
print('✅ Dependencies oʻrnatildi')

# System FFmpeg
os.system('apt-get install -y ffmpeg -qq')
print('✅ FFmpeg oʻrnatildi')

# Static ffmpeg (zaxira uchun)
STATIC_FFMPEG_URL = 'https://johnvansickle.com/ffmpeg/releases/ffmpeg-release-amd64-static.tar.xz'
if not os.path.exists('staticfiles/ffmpeg'):
    os.makedirs('staticfiles', exist_ok=True)
    os.system(f'wget -q {STATIC_FFMPEG_URL} -O /tmp/ffmpeg.tar.xz')
    os.system('tar -xf /tmp/ffmpeg.tar.xz -C /tmp/')
    bins = glob.glob('/tmp/ffmpeg-*-amd64-static/ffmpeg')
    if bins:
        shutil.copy(bins[0], 'staticfiles/ffmpeg')
        os.chmod('staticfiles/ffmpeg', 0o755)
        print('✅ Static ffmpeg saqlandi')

clear_output()
print('✅ Setup tugadi — Hujayra 2 ga oʻting')
```

```python
import os

#@markdown ## ⚙️ Bot konfiguratsiyasi
#@markdown ---
#@markdown `.env` fayl mavjud bo'lsa, uning qiymatlari **avtomatik** ishlatiladi.
#@markdown Bo'sh qoldirsangiz ham `.env` dan o'qiydi. `.env` yo'q bo'lsa form qiymati ishlatiladi.
#@markdown ---

#@markdown **API ID** — my.telegram.org dan oling
API_ID = ""  #@param {type:"string"}

#@markdown **API HASH** — my.telegram.org dan oling
API_HASH = ""  #@param {type:"string"}

#@markdown **BOT TOKEN** — @BotFather dan oling
BOT_TOKEN = ""  #@param {type:"string"}

#@markdown **MongoDB URI** — bo'sh qolsa local JSON ishlatiladi
DB_URI = ""  #@param {type:"string"}

# ── .env fayldan o'qish (agar mavjud bo'lsa) ──────────────────────────────
def _load_dotenv(path=".env"):
    """Oddiy .env parser — python-dotenv shart emas."""
    result = {}
    if not os.path.exists(path):
        return result
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith("#"):
                continue
            if "=" not in line:
                continue
            key, _, val = line.partition("=")
            result[key.strip()] = val.strip().strip('"').strip("'")
    return result

_env = _load_dotenv(".env")

def _resolve(form_val, env_key):
    """
    Qiymat tanlash tartibi:
      1. Form bo'sh → .env dan ol
      2. Form to'ldirilgan → form qiymatini ishlatiladi
      3. Ikkalasi ham yo'q → bo'sh string
    """
    form_val = (form_val or "").strip()
    env_val  = _env.get(env_key, "").strip()
    # Form to'ldirilmagan bo'lsa .env ni afzal ko'rish
    return form_val if form_val else env_val

# ── Muhit o'zgaruvchilariga yozish ────────────────────────────────────────
final_api_id    = _resolve(API_ID,    "API_ID")
final_api_hash  = _resolve(API_HASH,  "API_HASH")
final_bot_token = _resolve(BOT_TOKEN, "BOT_TOKEN")
final_db_uri    = _resolve(DB_URI,    "DB_URI")

os.environ["API_ID"]    = final_api_id
os.environ["API_HASH"]  = final_api_hash
os.environ["BOT_TOKEN"] = final_bot_token
os.environ["DB_URI"]    = final_db_uri

# ── Natijani ko'rsatish ────────────────────────────────────────────────────
src = "(.env)" if _env else "(form)"
print(f"{'✅' if final_api_id    else '❌'} API_ID    : {'*' * len(final_api_id)    if final_api_id    else 'KIRITILMADI'} {src}")
print(f"{'✅' if final_api_hash  else '❌'} API_HASH  : {'*' * len(final_api_hash)  if final_api_hash  else 'KIRITILMADI'} {src}")
print(f"{'✅' if final_bot_token else '❌'} BOT_TOKEN : {'*' * len(final_bot_token) if final_bot_token else 'KIRITILMADI'} {src}")
if final_db_uri:
    print(f"✅ DB_URI   : {final_db_uri[:20]}... {src}")
else:
    print(f"⚠️  DB_URI kiritilmadi — local JSON fallback (sessions/local_db.json)")

missing = [k for k, v in [("API_ID", final_api_id), ("API_HASH", final_api_hash), ("BOT_TOKEN", final_bot_token)] if not v]
if missing:
    print(f"\n❌ Quyidagilar kiritilmagan: {', '.join(missing)}")
    print("   .env fayliga yozing yoki formga kiriting.")
else:
    print("\n✅ Config tayyor — Hujayra 3 ni ishga tushiring")
```
